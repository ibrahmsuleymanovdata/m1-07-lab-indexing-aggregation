```python
data = {
    'student_id': [f"S06{i}" for i in range(10)],
    'status': ['active', 'inactive', 'active', 'inactive', 'active', 
               'inactive', 'active', 'inactive', 'active', 'inactive']
}

roster = pd.DataFrame(data)
```

```python
merged_df = pd.merge(scores,roster, how = "left", on = "student_id")
```

```python
missing_status_count = merged_df['status'].isna().sum()

print(f"NAN values in status column: {missing_status_count}")
```

# Task 3

```python
avg_by_module=merged_df.groupby(["cohort","module"])['score'].mean().reset_index()
```

```python
avg_by_module
```

```python
avg_by_cohort = scores.groupby('cohort')['score'].mean().reset_index()
avg_by_cohort
```

```python
rolled_up = avg_by_module.groupby('cohort')['score'].mean()

print("avg with module:")
print(rolled_up)

print("\navg with cohort(original):")
print(avg_by_cohort.set_index('cohort')['score'])
```

# Task 4

```python
student_module_report = scores.pivot_table(index="student_id", columns="module", values="score",aggfunc="mean")
student_module_report.head()
```

```python
student_module_report.index
```

```python
scores["student_id"].unique() == student_module_report.index
```

```python
unique_students_count = scores['student_id'].nunique()
report_rows_count = student_module_report.shape[0]

unique_modules = sorted(scores['module'].unique())
report_columns = sorted(student_module_report.columns.tolist())

print(f"Original unique student count: {unique_students_count}")
print(f"Report row count: {report_rows_count}")

row_check = "Successful" if unique_students_count == report_rows_count else "Error!"
print(f"Validation 1 (Rows): {row_check}")

print("-" * 30)

print(f"Original modules: {unique_modules}")
print(f"Report columns: {report_columns}")
col_check = "Successful" if unique_modules == report_columns else "Error!"
print(f"Validation 2 (Columns): {col_check}")
```

# Task 5

```python
student_avgs = scores.groupby(['student_id', 'cohort'])['score'].mean().reset_index()
top_students = student_avgs.sort_values(['cohort', 'score'], ascending=[True, False]) \
                           .groupby('cohort') \
                           .head(3) \
                           .reset_index(drop=True)

top_students.columns = ['student_id', 'cohort', 'avg_score']


print("Top 3 Students by Cohort:")
print(top_students)
```
